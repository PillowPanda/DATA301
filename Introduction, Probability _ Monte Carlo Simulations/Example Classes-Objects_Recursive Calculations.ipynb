### A More Involving Example

The __init__ is a method that is used to initialize attributes for a new instance of a class.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Joe", 19)

print(p1.name)
```

Example: a class-based, iterative approach to generating the Fibonacci sequence:

```python
class Fib:
  def __init__(self):
    self.a = 0
    self.b = 1
  def update(self):
    self.a, self.b = self.b, self.a + self.b
    self.current = self.b
```

```python
f = Fib()
```

```python
f.update()
f.current
```

Another coding example involving iterators:

```python
class FibNumbers:
  def __iter__(self):
    self.a = 0
    self.b = 1
    return self

  def __next__(self):
    self.a, self.b = self.b, self.a + self.b
    return self.b
```

```python
myclass = FibNumbers()
myiter = iter(myclass)
```

```python
next(myiter)
```

The *self* parameter is a reference to the current instance of the class, and is used to access variables that belongs to the class.

It does not have to be named *self* , you can call it whatever you like, but it has to be the first parameter of any function in the class.

## Recursive Calculations

```python
def fibonacci_of(n):
     if n in {0, 1}:  # Base case
         return n
     return fibonacci_of(n - 1) + fibonacci_of(n - 2)  # Recursive case
```

```python
fibonacci_of(20)
```
