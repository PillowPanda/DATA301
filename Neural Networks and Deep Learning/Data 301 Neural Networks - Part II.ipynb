## <font color='blue' size=6pt> Optimization Algorithms for Backpropagation</font>

All the current optimization algorithms are based on a variant of gradient descent.

Let's denote the vector of weights at step $t$ by $w_t$ and the gradient of the objective function with respect to the weights by $g_t$. The idea is that the gradient descent algorithm updates the weights under the following principle:

$$\large w_t = w_{t-1} - \eta\cdot g_{t,t-1}$$

When the objective function (whose gradient with respect to the weights) is represented by $g_t$ and has multiple local minima, or it has a very shallow region containing the minima, the plain gradient descent algorithm may not converge to the position sought for. To remediate this deficiency research proposed alternatives by varying the way we evaluate the learning rate each step or how we compute the "velocity" for updating the weights:

<font color='green'>$$\Large w_t = w_{t-1} - \eta_t\cdot v_t$$

In the equation above, $\eta_t$ is an adaptive learning rate and $v_t$ a modified gradient.

## Dynamic Learning Rates

We can consider an exponential decay, such as

$$\large \eta_t = \eta_0 e^{-\lambda\cdot t}$$

or a polynomial decay

$$\large \eta_t = \eta_0 (\beta t+1)^{-\alpha}$$

## Momentum Gradient Descent

$$\Large g_{t,t-1} = \partial_w \frac{1}{|\text{B}_t|}\sum_{i\in \text{B}_t}f(x_i,w_{t-1})=\frac{1}{|\text{B}_t|}\sum_{i\in \text{B}_t}h_{i,t-1}$$

where

$$\large v_t = \beta v_{t-1} + g_{t,t-1}$$

and $\beta\in (0,1).$

For an explicit formula, we have

$$\large v_t = \sum_{\tau=0}^{t-1} \beta^{\tau}g_{t-\tau,t-\tau-1}$$


and

$$\large w_t = w_{t-1} - \alpha v_t$$

where

$$\large \alpha = \frac{\eta}{1-\beta}$$


## AdaGrad

$$\large s_t = s_{t-1} + g_{t}^2$$

and

$$\large w_t= w_{t-1} - \frac{\eta}{\sqrt{s_t+\epsilon}}\cdot g_t$$

## RMSProp

$$\large s_t = \gamma\cdot s_{t-1} + (1-\gamma)\cdot g_{t}^2$$

and

$$\large w_t= w_{t-1} - \frac{\eta}{\sqrt{s_t+\epsilon}}\cdot g_t$$

Thus, we have

$$\large s_t = (1-\gamma)\cdot (g_t^2+\gamma g_{t-1}^2+\gamma^2 g_{t-2} +\gamma^3 g_{t-2} + ... + \gamma^{\tau} g_0)$$

## ADAM (Adaptive Momentum Gradient Descent)

$$\large v_t = \beta_1  v_{t-1} +(1-\beta_1) g_t$$

and

$$\large s_t = \beta_2 s_{t-1} + (1-\beta_2) g_t^2 $$

We further consider

$$\large \hat{v}_t = \frac{v_t}{1-\beta_1}, \text{and  } \hat{s}_t = \frac{s_t}{1-\beta_2} $$


and

$$\large \hat{g}_{t} = \frac{\eta\cdot \hat{v}_t}{\sqrt{\hat{s}_t}+\epsilon}$$

The updates to the weights are implemented as follows

$$\large w_t = w_{t-1} - \hat{g}_t$$

## Example of a Neural Network for classification

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data= data.data,columns=data.feature_names)
df
```

Let's suppose we use all the different features.

```python
X = df.values
```

```python
y = data.target
```

```python
y
```

We need to provide the dimensionality of the input features

```python
ndim = X.shape[1]
```

```python
ndim
```

### Here we design a neural network for classification

```python
model = Sequential() # this is almost always the case
# then we start adding layers of artifical neurons
model.add(Dense(3,activation='gelu',use_bias=True,input_dim=ndim))# the first layer
#model.add(Dense(8,activation=ReLU(), kernel_initializer=initializers.RandomUniform(minval=-0.05, maxval=0.05, seed=123)))# the second layer
#model.add(Dense(4,activation=LeakyReLU(alpha=0.01))) # the third layer
model.add(Dense(1,activation='sigmoid')) # the output layer
```

```python
opt = Adam(learning_rate=0.1)
model.compile(loss='binary_crossentropy',optimizer=opt,metrics='accuracy') # it's important to pick the correct loss and the correct metric
```

```python
# we most likely need to standardize the data
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
# first let's use the whole data
Xs = scale.fit_transform(X)
```

```python
# here we train the Artificial Neural Network that we designed
model.fit(Xs,y,batch_size=64,epochs=80,validation_batch_size=0.2,verbose=1) # what is an epoch?
```

```python
model.evaluate(Xs,y)
```

```python
proba = model.predict(Xs) # these are probability values so the output is SOFT CLASSIFICATION
```

```python
proba[:20] # the probability of benign tumors for the first 10 patients
```

```python
y[:10]
```

```python
# to harden the soft classification you need the following
y_predicted = (model.predict(Xs) > 0.5).astype("int32").reshape(-1,1)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pd.DataFrame(data=confusion_matrix(y,y_predicted),columns = data.target_names,index=data.target_names)
```

Here we want to validate the proposed design of the Neural Network

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True,random_state=310)
```

```python
test_acc = []
for idxtrain, idxtest in skf.split(X,y):
  xtrain = X[idxtrain]
  # we should scale properly
  xtrain = scale.fit_transform(xtrain)
  xtest = X[idxtest]
  xtest = scale.transform(xtest)
  # so we got the input features scaled correctly for cross-validation
  ytrain = y[idxtrain]
  ytest = y[idxtest]
  # the design we want to validate
  model = Sequential() # this is almost always the case
  # then we start adding layers of artifical neurons
  model.add(Dense(16,activation=LeakyReLU(alpha=0.1),input_dim=ndim))# the first layer
  model.add(Dense(8,activation=ReLU()))# the second layer
  model.add(Dense(12,activation=LeakyReLU(alpha=0.05))) # the third layer
  model.add(Dense(1,kernel_regularizer=l2(0.01),activation='sigmoid')) # the output layer
  opt = Adam(learning_rate=0.15)
  model.compile(loss='binary_crossentropy',optimizer=opt,metrics='accuracy')
  model.fit(xtrain,ytrain, verbose=0)
  #collect the evaluation on the test sets
  test_acc.append(model.evaluate(xtest,ytest)[1])
```

```python
np.mean(test_acc)
```

```python
model.evaluate(Xs,y)
```

```python
proba = model.predict(Xs)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
fpr , tpr , thresholds = roc_curve ( y , proba)
```

```python
def plot_roc_curve(fpr,tpr):
  plt.figure(figsize=(6,6))
  plt.plot(fpr,tpr,lw=2)
  plt.axis([0,1,0,1])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.show()
```

```python
plot_roc_curve(fpr,tpr)
```

```python
roc_auc_score(y,proba)
```
